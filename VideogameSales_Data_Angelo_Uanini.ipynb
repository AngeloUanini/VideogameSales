# 1) Que plataformas tuvieron las mayores ventas?


---



```python
# Agrupar por plataforma y sumar las ventas globales
platform_sales = df.groupby('Platform')['Global_Sales'].sum().reset_index()

# Ordenar las plataformas por ventas
platform_sales = platform_sales.sort_values(by='Global_Sales', ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=platform_sales, x='Platform', y='Global_Sales', palette='viridis')
plt.title('Ventas Globales por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (millones)')
plt.xticks(rotation=45)
plt.show()
```

Podemos ver que en las ventas historicas la PS2 es la que mas ventas tiene por bastante con mas de 1200 millones de copias de videojuegos vendidas, esto se puede deber a que esta plataforma tuvo muchisimo exito en su momento, tanto que no se dejo de fabricar hasta la salida de la PS4.

# 2) Que regiones tuvieron mas ventas?


---



```python
# Calcular las ventas totales por región
region_sales = {
    'North America': df['NA_Sales'].sum(),
    'Europe': df['EU_Sales'].sum(),
    'Japan': df['JP_Sales'].sum(),
    'Other': df['Other_Sales'].sum(),
}

# Crear un DataFrame a partir del diccionario
region_sales_df = pd.DataFrame(region_sales.items(), columns=['Region', 'Sales'])

# Crear el gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(region_sales_df['Sales'], labels=region_sales_df['Region'], autopct='%1.1f%%', startangle=140)
plt.title('Participación de Ventas Globales por Región')
plt.axis('equal')  # Para que el pastel sea un círculo
plt.show()
```

Podemos ver que la region que mas ventas tiene es Norte America, esto se puede deber al exito de las PlayStation en general ya que eran muy exitosas en norte america, ademas de que muchos exitos de Japon son muy jugados en Norte America.

# 3) Cómo han cambiado las ventas de videojuegos a lo largo del tiempo?


---



```python
# Agrupar por año y sumar las ventas globales
yearly_sales = df.groupby('Year')['Global_Sales'].sum().reset_index()

# Crear el gráfico de líneas
plt.figure(figsize=(14, 7))
sns.lineplot(data=yearly_sales, x='Year', y='Global_Sales', marker='o', color='b')
plt.title('Ventas Globales de Videojuegos a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (millones)')
plt.grid(True)
plt.xticks(yearly_sales['Year'], rotation=45)
plt.show()
```

Podemos ver que el Boom de ventas se vio a partir del año 2000. Esto se puede deber a que en esta epoca salieron todas las consolas historicas tales como la PS2, Xbox, GameCube, Xbox 360, PS3, Wii, Game Boy Advance, Nintendo DS y la PSP.

Podemos ver la caida de todas las ventas en general en el 2017, esto se puede deber a que se dejaron de hacer copias de los videojuegos en formato disco, cortucho, etc (formato fisico) y empezaron a venderse en formato digital, sea en las distintas plataformas de ventas o individualmente. Ademas de los avances de la pirateria dentro de los videojuegos por lo cual se podian jugar sin comprarlos. Lo mismo ocurrio con la aparicion de los juegos gratuitos, esto atribuyo a que disminuyeran las ventas.


# 4) Existe una relación entre los géneros de videojuegos y las plataformas en términos de ventas globales?


---



```python
# Agrupar las ventas por género y plataforma
heatmap_data = df.groupby(['Genre', 'Platform'])['Global_Sales'].sum().unstack()

# Crear el heatmap
plt.figure(figsize=(20, 14))  # Tamaño de la figura
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='.1f',  # Formato con un decimal
    cmap='YlGnBu',
    linewidths=2.5,  # Aumentar el ancho de las líneas para crear más separación
    annot_kws={"size": 10},  # Tamaño de las anotaciones
    square=True,  # Asegurar que las celdas sean cuadradas
    cbar_kws={
        'label': 'Ventas Globales (millones)',  # Etiqueta de la barra de color
        'shrink': 0.8,  # Reducción del tamaño de la barra de color
        'orientation': 'vertical'  # Mantener la barra vertical
    }
)
plt.title('Ventas Globales por Género y Plataforma', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Género', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.show()
```

Podemos observar que la relacion se encuentra mayormente en los juegos de deportes, accion, shooter, roleplay y racing.
Estos generos a dia de hoy siguen siendo los mas populares sin importar la plataforma donde se jueguen.
Podemos destacar puntos especificos como los deportes en la Wii y la PS2, los de accion en PS2, PS3 Y X360, y shooters en X360.

# 5) Cuáles son los géneros de videojuegos más populares globalmente y por región?


---



```python
# Agrupar las ventas por género globalmente
genre_global_sales = df.groupby('Genre').sum()[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sort_values('Global_Sales', ascending=False)

# Crear el gráfico de barras para las ventas globales por género
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_global_sales.index, y=genre_global_sales['Global_Sales'], palette='Set2')
plt.title('Ventas Globales por Género', fontsize=16)
plt.xlabel('Género', fontsize=14)
plt.ylabel('Ventas Globales (millones)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

Podemos ver que el genero que mas se ah vendido es Accion, seguido por deportes y luego por shooter. Tal como habiamos visto que eran los mas exitosos en el HeatMap.

```python
# Gráfico de barras apiladas para mostrar las ventas desglosadas por región
genre_global_sales[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Accent')
plt.title('Ventas por Género Desglosadas por Región', fontsize=16)
plt.xlabel('Género', fontsize=14)
plt.ylabel('Ventas por Región (millones)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Región")
plt.show()
```

Y podemos ver que las ventas por genero desglosadas por region estan relacionadas con la cantidad de ventas totales por region.
Lo interesante que se puede ver es que Japon no sigue la misma tendencia que las otras regiones ya que el genero mas vendido es Role-Playing.

# **6) Cuál es la plataforma más exitosa en términos de ventas para un género específico?**


---



```python
# Filtrar por un género específico (por ejemplo, "Action")
genre_specific = 'Action'
genre_platform_sales = df[df['Genre'] == genre_specific].groupby('Platform').sum()[['Global_Sales']].sort_values('Global_Sales', ascending=False)

# Crear un gráfico de barras para ver las plataformas más exitosas en ese género
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_platform_sales.index, y=genre_platform_sales['Global_Sales'], palette='Set3')
plt.title(f'Ventas Globales en el Género "{genre_specific}" por Plataforma', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Ventas Globales (millones)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

En este caso usamos el genero mas vendido para ver en que plataforma fue mas exitoso. Podemos ver que a pesar de que la PS2 es la plataforma con mas ventas en cantidad, la PS3 es la que tiene mas ventas en el genero action.

# **7) Qué géneros han perdido o ganado popularidad a lo largo de los años?**


---



```python
# Agrupar las ventas por año y género
genre_year_sales = df.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack()

# Crear el gráfico de líneas para cada género
plt.figure(figsize=(14, 8))
genre_year_sales.plot(kind='line', linewidth=2, marker='o', figsize=(14, 8), colormap='tab10')

# Ajustar el gráfico
plt.title('Evolución de las Ventas Globales por Género a lo Largo de los Años', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Ventas Globales (millones)', fontsize=14)
plt.legend(title="Género", loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Podemos ver que el pico de los juegos de deportes, accion, shooter y misc fue durante los años 2000, que coincide con el pico de ventas en estas consolas como PS2, PS3, Wii, etc. Asi mismo a partir del 2010 la mayoria de las ventas de estos cayo drasticamente.

# **8) Qué correlación existe entre las ventas en diferentes regiones?**


---



```python
# Seleccionar las columnas de ventas por región
regional_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

# Calcular la matriz de correlación
correlation_matrix = regional_sales.corr()

# Crear un heatmap de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Matriz de Correlación de Ventas Regionales')
plt.show()
```

Podemos ver que hay muchas correlacion entre NA Y EU y el que mejor correlacion tiene con el resto de regiones es EU, la correlacion de Japon con el resto es bastante baja por lo que tiene que ver con una preferencia en los videojuegos mas localizada a diferencia de las otras regiones.
Otro punto a tener en cuenta es la economia ya que hay paises donde tienen menos accesos a estos juegos o consolas, sea por precio o por exclusividad de las empresas, esto se ve reflejado en la correlacion de Otras ventas con el resto de regiones.
