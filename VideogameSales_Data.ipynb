```python
# Imputar la mediana en la columna "Year"
median_year = df['Year'].median()
df['Year'] = df['Year'].fillna(median_year)
```

```python
# Imputar "Unknown" en la columna "Publisher"
df['Publisher'] = df['Publisher'].fillna('Unknown')
```

```python
# Convertir la columna "Year" a tipo entero
df['Year'] = df['Year'].astype(int)
```

```python
# Verificar la limpieza
print(df.isnull().sum())
```

```python
# Agrupar por plataforma y sumar las ventas globales
platform_sales = df.groupby('Platform')['Global_Sales'].sum().reset_index()

# Ordenar las plataformas por ventas
platform_sales = platform_sales.sort_values(by='Global_Sales', ascending=False)

# Crear el grfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=platform_sales, x='Platform', y='Global_Sales', palette='viridis')
plt.title('Ventas Globales por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (millones)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar por ao y sumar las ventas globales
yearly_sales = df.groupby('Year')['Global_Sales'].sum().reset_index()

# Crear el grfico de lneas
plt.figure(figsize=(14, 7))
sns.lineplot(data=yearly_sales, x='Year', y='Global_Sales', marker='o', color='b')
plt.title('Ventas Globales de Videojuegos a lo Largo del Tiempo')
plt.xlabel('Ao')
plt.ylabel('Ventas Globales (millones)')
plt.grid(True)
plt.xticks(yearly_sales['Year'], rotation=45)
plt.show()
```
